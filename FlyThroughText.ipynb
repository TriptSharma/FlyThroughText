```python
from PIL import Image
from lang_sam import LangSAM

model = LangSAM()
image_pil = Image.open("/content/waldo_kitchen/images/frame_00001.jpg").convert("RGB")
text_prompt = "utensils toasters"
masks, boxes, phrases, logits = model.predict(image_pil, text_prompt)
```
